```
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```
columns = ['AirSpace Class', 'From_City' , 'To_City' , 'Price' , 'Aircraft Model' , 'Engine Model']
flight_file = pd.read_csv('https://raw.githubusercontent.com/13020363/UTS_ML2019_ID13020363/master/Flight.csv', names=columns)
```

```
cvTn_names = ['From_City', 'To_City', 'Aircraft Model', 'Engine Model']

for cvTn_name in cvTn_names:
  new_name = cvTn_name + '_category'
  flight_file[cvTn_name] = flight_file[cvTn_name].astype('category')
  flight_file[new_name] = flight_file[cvTn_name].cat.codes
```

```
# Visualizing 5-D mix data using bubble charts
# leveraging the concepts of hue, size and depth
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
t = fig.suptitle('Wine Residual Sugar - Alcohol Content - Acidity - Total Sulfur Dioxide - Type', fontsize=14)

xs = list(flight_file['From_City_category'])
ys = list(flight_file['To_City_category'])
zs = list(flight_file['Aircraft Model_category'])
  
data_points = [(x, y, z) for x, y, z in zip(xs, ys, zs)]

ss = list(flight_file['Price'])
colors = ['red' if wt == ' A' else 'yellow' if wt == ' B' else 'green' if wt == ' C' else 'black' if wt == ' D' else 'purple' if wt == ' E' else 'blue' for wt in list(flight_file['AirSpace Class'])]

for data, color, size in zip(data_points, colors, ss):
    x, y, z = data
    ax.scatter(x, y, z, alpha=0.4, c=color, edgecolors='none', s=size)

ax.set_xlabel('From_city')
ax.set_ylabel('To_City')
ax.set_zlabel('AirSpace Class')
```
